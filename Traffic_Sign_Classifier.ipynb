### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.

import tensorflow as tf

# random saturation change
r_saturation_X = tf.image.random_saturation(X_train, lower=0.4, upper=0.7)
processed_X_train = np.concatenate((X_train, r_saturation_X), axis=0)
processed_y_train = np.concatenate((y_train, y_train), axis=0)

### Plot images with it label
plt.figure(figsize=(15,20))
for i in range(0, n_classes):
    plt.subplot(9, 5, i+1)
    plt.subplots_adjust(wspace=0.2)
    x_selected = processed_X_train[processed_y_train == i]
    random_selected_index = random.randint(x_selected.shape[0]/2, x_selected.shape[0]-1)
    plt.imshow(x_selected[random_selected_index, :, :, :])
    plt.title(all_labels[i], fontsize=8)
    plt.axis('off')
plt.show()

# normalization(standardization): avoid out of memory
std_processed_X_train = tf.image.per_image_standardization(tf.convert_to_tensor(processed_X_train, np.float32))
std_X_valid = tf.image.per_image_standardization(tf.convert_to_tensor(X_valid, np.float32))
std_X_test = tf.image.per_image_standardization(tf.convert_to_tensor(X_test, np.float32))

print("Number of preprocessed training examples =", len(processed_X_train))
````

### Model Architecture

````python
### Define your architecture here.

nn_model = tf.keras.models.Sequential()

# Layer : Convolutional. Input:32x32x3, Output:30x30x6. (filters=6, kernel_size=(3, 3))
nn_model.add(tf.keras.layers.Conv2D(6, (3, 3), padding='VALID', activation='relu', input_shape=(32, 32, 3)))

# Layer : Dropout.
nn_model.add(tf.keras.layers.Dropout(0.2))

# Layer : Pooling. Input:30x30x6, Output:28x28x6. (pool_size=(3, 3), strides=(1, 1))
nn_model.add(tf.keras.layers.MaxPooling2D((3, 3), (1, 1), padding='VALID'))

# Layer : Convolutional. Input:28x28x6, Output:24x24x11. (filters=14, kernel_size=(5, 5))
nn_model.add(tf.keras.layers.Conv2D(14, (5, 5), padding='VALID', activation='relu'))

# Layer : Dropout.
nn_model.add(tf.keras.layers.Dropout(0.3))

# Layer : Pooling. Input:24x24x11, Output:12x12x11. (pool_size=(2, 2), strides=(2, 2))
nn_model.add(tf.keras.layers.MaxPooling2D((2, 2), (2, 2), padding='VALID'))

# Layer : Convolutional. Input:12x12x11. Output:6x6x18. (filters=22, kernel_size=(7, 7))
nn_model.add(tf.keras.layers.Conv2D(22, (7, 7), padding='VALID', activation='relu'))

# Layer : Dropout
nn_model.add(tf.keras.layers.Dropout(0.5))

# Layer : Pooling. Input:6x6x18, Output:5x5x22. (pool_size(2, 2), strides=(1, 1))
nn_model.add(tf.keras.layers.MaxPooling2D((2, 2), (1, 1), padding='VALID'))

# Layer : Fully Connected. Input:5x5x22, Output:550.
nn_model.add(tf.keras.layers.Flatten())

# Layer : Fully Connected. Input:550, Output:150.
nn_model.add(tf.keras.layers.Dense(150, kernel_regularizer=tf.keras.regularizers.l2(0.003), activation='relu'))

# Layer : Fully Connected. Input:150, Output:100.
nn_model.add(tf.keras.layers.Dense(100, kernel_regularizer=tf.keras.regularizers.l2(0.001), activation='relu'))

# Layer : Fully Connected. Input:100, Output:43.
nn_model.add(tf.keras.layers.Dense(43, activation='softmax'))

# show model summary
nn_model.summary()
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.

# compile the model
nn_model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# train the model
history = nn_model.fit(std_processed_X_train, processed_y_train, epochs=25, batch_size=256, 
                       validation_data=(std_X_valid, y_valid), verbose=2)
````

````python
# learning history result
plt.figure(figsize=(15, 2))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'valid'], loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'valid'], loc='upper right')
plt.show()
````

````python
# evaluate the model
test_loss, test_acc = nn_model.evaluate(std_X_test, y_test, verbose=0)
print('test_loss    : ' + str(test_loss))
print('test_accuracy: ' + str(test_acc))
````

````python
### Summary of prediction for test set

## function for summary of prediction
def plot_image_with_label(img, predicted_label_idx, predicted_label, true_label_idx, true_label, prediction_prob):
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap=plt.cm.binary)

    if predicted_label_idx == true_label_idx:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{}\n {:2.0f}% \n({})".format(predicted_label, prediction_prob*100, true_label), color=color)
    

def plot_top5_value_array(prediction_array, all_labels):

    prediction_top5_idx = (-prediction_array).argsort()[:5]
    prediction_top5_array = prediction_array[prediction_top5_idx]
    top5_label = [all_labels[i] for i in prediction_top5_idx]

    plt.grid(False)
    plt.xticks(range(5), top5_label, rotation=90, size=9, wrap=True, ha='left')
    plt.yticks([])
    thisplot = plt.bar(range(5), prediction_top5_array, color='#777777', width=0.3)
    for x in range(5):
        plt.text(x-0.35, -0.05, "{:2.1f}%".format(prediction_top5_array[x]*100), rotation=90)
    plt.ylim([0, 1])

    thisplot[0].set_color('cyan')
    
    
def prediction_analyze(input_images, test_images, test_label_id, num_rows, num_cols, random_choice=True):
    # load model prediction result for test set
    prediction = nn_model.predict(tf.convert_to_tensor(input_images, np.float32))

    # calculate number of output image
    num_images = num_rows * num_cols

    # plot random select result
    indexes = np.arange(len(test_label_id))
    if random_choice:
        indexes = np.random.choice(indexes, num_images, replace=False)
    plt.figure(figsize=(3*2*num_cols, 4*num_rows))
    plt.subplots_adjust(hspace=1.0, wspace=0.1)
    count = 0
    for i in indexes:
        img = test_images[i]
        prediction_array = prediction[i]
        prediction_prob = np.max(prediction_array)
        predicted_label_idx = np.argmax(prediction_array)
        predicted_label = all_labels[predicted_label_idx]
        true_label_idx = test_label_id[i]
        true_label = all_labels[true_label_idx]

        # plot random choiced image
        plt.subplot(num_rows, 2*num_cols, 2*count+1)
        plot_image_with_label(img, predicted_label_idx, predicted_label, true_label_idx, true_label, prediction_prob)

        # plot top5 value
        plt.subplot(num_rows, 2*num_cols, 2*count+2)
        plot_top5_value_array(prediction_array, all_labels)

        count += 1
    plt.show()
    

# analyze prediction result of test set
prediction_analyze(std_X_test, X_test, y_test, num_rows=5, num_cols=3)
````
